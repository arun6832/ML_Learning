```python
import re

# Example text containing GST numbers
texts = [
    "The GST number of company ABC is 12ABCDE3456F1Z7",
    "The GST number of company XYZ is 34FGHI6789J2Y8"
]

# Define a regex pattern to match GST numbers
gst_pattern = r"\b\d{2}[A-Z0-9]{10,15}\b"

# Extract GST numbers from each text
all_gst_numbers = []
for text in texts:
    gst_numbers = re.findall(gst_pattern, text)
    all_gst_numbers.extend(gst_numbers)

# Print extracted GST numbers
print("Extracted GST Numbers:", all_gst_numbers)
```
