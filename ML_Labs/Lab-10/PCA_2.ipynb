```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('cancerAll.xlsx')
df
```

```python
from sklearn.preprocessing import LabelEncoder

X = df.drop(['diagnosis','ID'],  axis=1)
y = df['diagnosis']

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
y = pd.DataFrame(y_encoded)
y = np.array(y)

feature_col = X.columns
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_np = scalar.fit_transform(X)
X_scalar = pd.DataFrame(X_np)
```

```python
# Feature importance 
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(X_scalar, y)
feature_importance = reg.feature_importances_
sort_indeces = np.argsort(feature_importance)[::-1]

sorted_importance = feature_importance[sort_indeces]
print(sorted_importance)
featuer_names = [f'Feature {i+1}' for i in range(X_scalar.shape[1])]
plt.bar(range(X_scalar.shape[1]), sorted_importance, align="center")
plt.xticks(range(X_scalar.shape[1]), [featuer_names[i] for i in sort_indeces], rotation=45, ha="right")
```
