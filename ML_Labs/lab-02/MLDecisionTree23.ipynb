### Some of the main parameters are performed below:

##### Splitter

```python
# splitter: It is one of the main paramenetr in the decision tree classifier. 
# It is used to select the splits for each of the nodes in the dtree. 
# There are 2 values for the parameter, one which will choose the 'best' split and the another one is choosed at 'random'.
# Initially it will be set into the 'best' splitting if we are not mentioning it.

features = ['Age', 'Experience', 'Rank', 'Nationality']
print(features)
X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier(criterion='entropy', splitter="random")
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

##### Max_depth

```python
#max_depth: The maximum depth of the tree. 
# If not specified or None, the tree is expanded until all leaves are pure or contain less than min_samples_split samples.

features = ['Age', 'Experience', 'Rank', 'Nationality']
print(features)
X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier(criterion='entropy', splitter="best", max_depth=1)
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

##### Random_state

```python
# random_state: This parameter allows you to control the randomness involved in certain operations, making the results reproducible.

features = ['Age', 'Experience', 'Rank', 'Nationality']
print(features)
X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier(criterion='entropy', splitter="best", max_depth=1, random_state=12)
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

## Advantages and Disadvantages of Decision Tree

#### Advantages:
<ul>
<li> Dicision tree requires less efforts for data preparation during pre-processing</li>
<li> It does not require normalization of the data.</li>
<li> It doesnot require scaling of data as well. </li>
<li> Missing values doesnot affect the performance of the tree</li>
<li> It is very easy to explain </li>
</ul>

#### Disadvantages:
<ul>
<li> A small change in the data can cause a big drastic change in the structure of the decision tree. </li>
<li> Sometimes the calculations may be more complex than other algorithms. </li>
<li> It often involves hight time to train the data </li>
<li> Training in Decision tree may cause more expensive in terms of complexity and time taken is more. </li>
<li> This algorithm is inadequate for applying regression and predicting continuous values. </li>
</ul>

## Some learnings on Decision Tree



<b>1. Introductrion to Decision Tree</b>

<ul>
<li>A supervised machine learning algorithm that can be applied to regression and classification problems is the decision tree.</li>
<li>Recursively dividing the dataset into subsets according to the most important feature at each node is how it operates.</li>
</ul>

<b>2. Components of Decision Tree</b>
<ol>
    <li><b>Root Node:</b> The topmost node in the Decision Tree</li>
    <li><b>Internal Node:</b> The node thatrepresent a decision.</li>
    <li><b>Leaf Node:</b> erminal nodes that provide the final output</li>
</ol>
<b>3. Splitting Creteria</b>
<ul>
    <li>Decision trees use various criteria to determine how to split the data at each node. Common criteria include Gini impurity and information gain (entropy).</li>
    <li>The goal is to create pure nodes with samples belonging to a single class.</li>
</ul>
<b>4. Some important terms</b>
<ul>
    <li><b>Entropy: </b>Measures the randomness or disorder in a set of samples. It is minimized when all samples belong to a single class.</li>
    <li><b>Gini impurity: </b> Measures the probability of incorrectly classifying a randomly chosen element. </li>
</ul>
